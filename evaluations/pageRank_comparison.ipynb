```python
basic_pagerank_results = []
pagerank_backflow_results = []
pagerank_selfflow_results = []
pagerank_norm_results = []
pagerank_norm_backflow_results = []
pagerank_norm_selfflow_results = []
pagerank_selfflow_backflow_results = []
postrank_results = []
```

```python
reverse = True
mapping = dict(zip(DG, range(0, DG.number_of_nodes())))
relabelled_DG = nx.relabel_nodes(DG, mapping)
if reverse == True:
    relabelled_DG = relabelled_DG.reverse()
```

```python
for i in range(10):
    reduced_graph, removed_edges = reduce_graph(relabelled_DG, perc_edge_del=0.2)

    print("basic_pagerank")
    basic_pagerank = generalised_page_rank(reduced_graph, self_loop_weight = 0, alpha = 1,
                               page_size = 0, end_normalise = False, arrow_dir_powerful = False,
                               backflow = False, backflow_const = 0,
                               min_iter = min_iter, print_rate = print_rate, max_iter = max_iter,
                               cut_off_change = cut_off_change)

    print("pagerank_backflow")
    pagerank_backflow = generalised_page_rank(reduced_graph, self_loop_weight = 0, alpha = 1,
                               page_size = 0, end_normalise = False, arrow_dir_powerful = False,
                               backflow = True, backflow_const = 0.2,
                               min_iter = min_iter, print_rate = print_rate, max_iter = max_iter,
                               cut_off_change = cut_off_change)

    print("pagerank_selfflow")
    pagerank_selfflow = generalised_page_rank(reduced_graph, self_loop_weight = 1, alpha = 1,
                               page_size = 1, end_normalise = False, arrow_dir_powerful = False,
                               backflow = False, backflow_const = 0,
                               min_iter = min_iter, print_rate = print_rate, max_iter = max_iter,
                               cut_off_change = cut_off_change)

    print("pagerank_norm")
    pagerank_norm = generalised_page_rank(reduced_graph, self_loop_weight = 0, alpha = 1,
                               page_size = 0, end_normalise = True, arrow_dir_powerful = False,
                               backflow = False, backflow_const = 0,
                               min_iter = min_iter, print_rate = print_rate, max_iter = max_iter,
                               cut_off_change = cut_off_change)

    print("pagerank_norm_backflow")
    pagerank_norm_backflow = generalised_page_rank(reduced_graph, self_loop_weight = 0, alpha = 1,
                               page_size = 0, end_normalise = True, arrow_dir_powerful = False,
                               backflow = True, backflow_const = 0.2,
                               min_iter = min_iter, print_rate = print_rate, max_iter = max_iter,
                               cut_off_change = cut_off_change)

    print("pagerank_norm_selfflow")
    pagerank_norm_selfflow = generalised_page_rank(reduced_graph, self_loop_weight = 1, alpha = 1,
                               page_size = 1, end_normalise = True, arrow_dir_powerful = False,
                               backflow = False, backflow_const = 0,
                               min_iter = min_iter, print_rate = print_rate, max_iter = max_iter,
                               cut_off_change = cut_off_change)

    print("pagerank_selfflow_backflow")
    pagerank_selfflow_backflow = generalised_page_rank(reduced_graph, self_loop_weight = 1, alpha = 1,
                               page_size = 1, end_normalise = False, arrow_dir_powerful = False,
                               backflow = True, backflow_const = 0.2,
                               min_iter = min_iter, print_rate = print_rate, max_iter = max_iter,
                               cut_off_change = cut_off_change)

    print("postrank")
    postrank = generalised_page_rank(reduced_graph, self_loop_weight = 1, alpha = 1,
                               page_size = 1, end_normalise = True, arrow_dir_powerful = False,
                               backflow = True, backflow_const = 0.2,
                               min_iter = min_iter, print_rate = print_rate, max_iter = max_iter,
                               cut_off_change = cut_off_change)


    basic_pagerank_results.append(perc_correct(removed_edges, basic_pagerank))
    pagerank_backflow_results.append(perc_correct(removed_edges, pagerank_backflow))
    pagerank_selfflow_results.append(perc_correct(removed_edges, pagerank_selfflow))
    pagerank_norm_results.append(perc_correct(removed_edges, pagerank_norm))
    pagerank_norm_backflow_results.append(perc_correct(removed_edges, pagerank_norm_backflow))
    pagerank_norm_selfflow_results.append(perc_correct(removed_edges, pagerank_norm_selfflow))
    pagerank_selfflow_backflow_results.append(perc_correct(removed_edges, pagerank_selfflow_backflow))
    postrank_results.append(perc_correct(removed_edges, postrank))
```

```python
print("basic_pagerank:", mean(basic_pagerank_results))
print("pagerank_backflow:", mean(pagerank_backflow_results))
print("pagerank_selfflow:", mean(pagerank_selfflow_results))
print("pagerank_norm:", mean(pagerank_norm_results))
print("pagerank_norm_backflow:", mean(pagerank_norm_backflow_results))
print("pagerank_norm_selfflow:", mean(pagerank_norm_selfflow_results))
print("pagerank_selfflow_backflow:", mean(pagerank_selfflow_backflow_results))
print("postrank:", mean(postrank_results))
```
